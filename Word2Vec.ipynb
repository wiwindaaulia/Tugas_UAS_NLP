```python
import os

import pandas as pd
from tqdm import tqdm

from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
```

Prepare Corpus

```python
# Baca file CSV
file_path = '/content/mobil_listrik.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
print(df.head())  # Cek beberapa baris pertama
print(df.info())  # Lihat tipe data kolom
```

```python
# Pastikan tidak ada NaN dan semua data bertipe string
df['text_cleaning'] = df['text_cleaning'].fillna('').astype(str)
```

```python
sentences = [text.lower().split() for text in tqdm(df['text_cleaning'])]
print(sentences[:5])
```

#Train Word2Vec Model

```python
model = Word2Vec(sentences, vector_size=128, window=5, min_count=3, workers=4, epochs=1000, sg=0, hs=0)
```

#Save

```python
os.makedirs("model/w2v/", exist_ok=True)
```

```python
model.save("model/w2v/mobil_listrik.w2v")
```

#load

```python
model = Word2Vec.load("model/w2v/mobil_listrik.w2v")
```

#Continue training

```python
from re import split
"saran sih bikin harga ionic sama kayak brio insya alloh laris manis".lower().split()
```

```python
contoh_data = [
    ['saran','sih','bikin','harga','ionic','sama','kayak','brio','insya','alloh','laris','manis'],
    ["baik", "kualitas", "kembang", "dulu", "baik", "kualitas", "motor", "motor", "pabrikan", "jepang"]
]
```

```python
model.train(contoh_data, total_examples=len(contoh_data), epochs=1)
```

```python
model.save("model/w2v/mobil_listrik.w2v")
```

##higher order visualiization

```python
from umap import UMAP
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
X = UMAP().fit_transform(w2v.vectors)
```

```python
df =pd.DataFrame(X, columns=["umap1", "umap2"])
df["text"] = w2v.index_to_key
```

```python
df
```

```python
import plotly.express as px

fig = px.scatter(df, x="umap1", y="umap2", text="text")
fig.update_traces(textposition="top center")
fig.update_layout(
    height=800,
    title_text="Reduced Word2Vec Visualization"
)

fig.show()
```
